# TRADITIONAL CODING 

```python
def my_add(a, b):
    result = a + b
    print(f"{a} + {b} = {result}")
    return result

my_add(0, 1)
```

# Using Reduce()

```python
from functools import reduce
numbers = [0, 1, 2, 3, 4]
reduce(my_add, numbers)
```

# Using Optinal Parameter

```python
numbers = [0, 1, 2, 3, 4]
reduce(my_add, numbers, 100)
```

# Not Support Blank Object

```python
from functools import reduce
# Using an initializer value
reduce(my_add, [], 0)  # Use 0 as return value

# Using no initializer value
reduce(my_add, []) 
```

# Finding the Minimum and Maximum Value

```python
# * is iterable unpacking operator
numbers = [3, 5, 2, 4, 7, 1]

# Minimum
min_value, *rest = numbers
print("Before Calculation Min Value:",min_value)
print("Before Calculation rest Value:",rest)
for num in rest:
    if num < min_value:
        min_value = num

print("Min Value:",min_value)


# Maximum
max_value, *rest = numbers
print("Before Calculation Max Value:",max_value)
print("Before Calculation rest Value:",rest)
for num in rest:
    if num > max_value:
        max_value = num

print("Max Value:",max_value)
```

```python
from functools import reduce

# Minimum
def my_min_func(a, b):
    return a if a < b else b

# Maximum
def my_max_func(a, b):
    return a if a > b else b

numbers = [3, 5, 2, 4, 7, 1]

print("Min After Reduce:",reduce(my_min_func, numbers))

print("Max After Reduce:",reduce(my_max_func, numbers))
```

```python
from functools import reduce

numbers = [3, 5, 2, 4, 7, 1]

# Minimum
print("Min Using Reducing:",reduce(lambda a, b: a if a < b else b, numbers))

# Maximum
print("Max Using Reducing:",reduce(lambda a, b: a if a > b else b, numbers))
```

```python
from functools import reduce

# a 2D list 
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]] 

# using reduce function for flattening a 2D list
print(reduce(lambda x, y: x + y, matrix))

# using list comprehension
print([i for row in matrix for i in row])
```
